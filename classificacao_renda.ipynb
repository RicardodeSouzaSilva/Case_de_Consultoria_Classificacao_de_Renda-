```python
import importlib    # Importa o m√≥dulo 'importlib', que permite importar bibliotecas dinamicamente em tempo de execu√ß√£o
import subprocess   # Importa o m√≥dulo 'subprocess' para executar comandos externos, como instalar pacotes via pip
import sys          # Importa o m√≥dulo 'sys' para acessar o interpretador Python atual e seu execut√°vel


# Dicion√°rio contendo os pacotes que devem ser verificados e instalados, se necess√°rio.
# A chave √© o nome usado para importar o m√≥dulo, e o valor √© o nome usado no pip para instala√ß√£o.
required_packages = {
    'sklearn': 'scikit-learn',     # sklearn √© importado como m√≥dulo, mas instalado com o nome scikit-learn
    'scipy': 'scipy',              # scipy √© tanto o nome de importa√ß√£o quanto de instala√ß√£o
    'pandas': 'pandas',            # pandas para an√°lise de dados tabulares
    'numpy': 'numpy',              # numpy para opera√ß√µes num√©ricas
    'matplotlib': 'matplotlib',    # matplotlib para visualiza√ß√µes
    'seaborn': 'seaborn',          # seaborn para gr√°ficos estat√≠sticos
    'missingno': 'missingno'       # missingno para visualiza√ß√£o de dados ausentes
}

# Fun√ß√£o respons√°vel por verificar se o pacote est√° instalado.
# Caso n√£o esteja, ela o instala automaticamente usando pip.
def check_and_install(package_name, install_name):
    try:
        # Tenta importar o pacote. Se for bem-sucedido, ele j√° est√° instalado.
        importlib.import_module(package_name)
        print(f"‚úÖ Biblioteca '{package_name}' j√° est√° instalada.")
    except ImportError:
        # Caso a importa√ß√£o falhe, significa que a biblioteca n√£o est√° instalada.
        # O c√≥digo abaixo usa o subprocess para executar o comando: python -m pip install <nome_do_pacote>
        print(f"‚ö†Ô∏è Biblioteca '{package_name}' n√£o encontrada. Instalando '{install_name}'...")
        subprocess.check_call([sys.executable, "-m", "pip", "install", install_name])
        print(f"‚úÖ Biblioteca '{install_name}' instalada com sucesso.")

# Loop que percorre cada item do dicion√°rio de pacotes.
# Para cada pacote, chama a fun√ß√£o de verifica√ß√£o e instala√ß√£o, se necess√°rio.
for package, install_name in required_packages.items():
    check_and_install(package, install_name)
```

```python
# Tabela com m√©dias por classe de trabalho
# Cria uma tabela resumo com as m√©dias por classe de trabalho
tabela_media_classe_trabalho = (
    dataset
        # 1) Agrupa o DataFrame pela vari√°vel qualitativa 'classe_trabalho'.
        # ------------------------------------------------------------------
        # `.groupby('classe_trabalho', as_index=False)` diz ao pandas:
        #   - Separe o DataFrame `dataset` em grupos, onde cada grupo cont√©m
        #     todas as linhas com o mesmo valor na coluna 'classe_trabalho'.
        #   - Exemplos de grupos: 'Private', 'Self-emp-not-inc', 'Local-gov',
        #     'State-gov', etc., dependendo das categorias presentes nos dados.
        #   - `as_index=False` garante que 'classe_trabalho' continue sendo
        #     uma coluna normal no DataFrame resultante (e n√£o vire √≠ndice).
        .groupby('classe_trabalho', as_index=False)

        # 2) Para cada grupo, calcula a M√âDIA das vari√°veis num√©ricas selecionadas.
        # ------------------------------------------------------------------
        # `.agg({...})` aplica fun√ß√µes de agrega√ß√£o para cada grupo criado no groupby.
        # Aqui usamos a fun√ß√£o 'mean' (m√©dia) em quatro colunas num√©ricas:
        #   - 'horas_semanais'       ‚Üí m√©dia de horas trabalhadas por semana
        #   - 'escolaridade_anos'    ‚Üí m√©dia de anos de estudo
        #   - 'ganho_capital'        ‚Üí m√©dia de ganhos de capital
        #   - 'perda_capital'        ‚Üí m√©dia de perdas de capital
        #
        # O resultado desta etapa √© um novo DataFrame onde:
        #   - Cada linha representa uma classe de trabalho.
        #   - As colunas num√©ricas representam as m√©dias dessas vari√°veis
        #     dentro de cada classe.
        .agg({
            'horas_semanais': 'mean',
            'escolaridade_anos': 'mean',
            'ganho_capital': 'mean',
            'perda_capital': 'mean'
        })

        # 3) Arredonda todas as colunas num√©ricas para 2 casas decimais.
        # ------------------------------------------------------------------
        # `.round(2)` aplica arredondamento com 2 casas decimais em todas as
        # colunas num√©ricas do DataFrame resultante da agrega√ß√£o.
        # Isso torna a tabela mais ‚Äúlimpa‚Äù e amig√°vel para leitura,
        # especialmente em relat√≥rios, gr√°ficos ou apresenta√ß√µes.
        .round(2)

        # 4) Renomeia as colunas de m√©dias para nomes mais descritivos.
        # ------------------------------------------------------------------
        # `.rename(columns={...})` altera os nomes das colunas resultantes,
        # deixando expl√≠cito que os valores s√£o M√âDIAS e n√£o observa√ß√µes individuais.
        # Assim, evitamos ambiguidade na interpreta√ß√£o dos resultados.
        .rename(columns={
            'horas_semanais': 'media_horas_semanais',        # m√©dia de horas semanais
            'escolaridade_anos': 'media_escolaridade_anos',  # m√©dia de anos de estudo
            'ganho_capital': 'media_ganho_capital',          # m√©dia de ganho de capital
            'perda_capital': 'media_perda_capital'           # m√©dia de perda de capital
        })
)

# Exibe o DataFrame final, que resume as m√©dias das vari√°veis num√©ricas
# para cada classe de trabalho.
tabela_media_classe_trabalho

```

```python
import math
import matplotlib.pyplot as plt
import seaborn as sns

# Configura o estilo visual padr√£o dos gr√°ficos.
# Aqui usamos "whitegrid", que coloca um fundo branco com grade,
# facilitando a leitura dos valores e a compara√ß√£o entre categorias.
sns.set(style="whitegrid")

# Lista com os nomes das colunas num√©ricas (m√©tricas) que ser√£o usadas nos gr√°ficos.
# Cada string aqui corresponde a uma coluna do DataFrame `tabela_media_classe_trabalho`,
# que cont√©m as m√©dias calculadas previamente por classe de trabalho.
metricas = [
    'media_horas_semanais',      # m√©dia de horas trabalhadas por semana
    'media_escolaridade_anos',   # m√©dia de anos de escolaridade
    'media_ganho_capital',       # m√©dia de ganho de capital
    'media_perda_capital'        # m√©dia de perda de capital
]

# Lista de t√≠tulos para cada subplot.
# A posi√ß√£o de cada t√≠tulo corresponde √† mesma posi√ß√£o da m√©trica em `metricas`.
# Assim, titulos[i] descreve o gr√°fico que est√° usando metricas[i].
titulos = [
    'M√©dia de horas semanais por classe de trabalho',
    'M√©dia de anos de escolaridade por classe de trabalho',
    'M√©dia de ganho de capital por classe de trabalho',
    'M√©dia de perda de capital por classe de trabalho'
]

# Define a quantidade de colunas no grid de subplots.
# Ou seja, ser√£o exibidos 2 gr√°ficos por linha.
col = 2

# Calcula dinamicamente a quantidade de linhas necess√°rias no grid.
# - len(metricas) √© o n√∫mero total de gr√°ficos que queremos criar.
# - Dividimos pelo n√∫mero de colunas (col) para saber quantas linhas precisamos.
# - math.ceil(...) arredonda para cima para garantir linhas suficientes,
#   mesmo se len(metricas) n√£o for m√∫ltiplo de col.
row = math.ceil(len(metricas) / col)

# Cria a figura (`fig`) e um array de eixos (`ax`) para os subplots.
# - `row` √© o n√∫mero de linhas de subplots.
# - `col` √© o n√∫mero de colunas de subplots.
# - `figsize=(22, 8)` define o tamanho da figura (largura x altura em polegadas),
#   para que os gr√°ficos fiquem grandes o suficiente e leg√≠veis.
fig, ax = plt.subplots(row, col, figsize=(22, 8))

# O objeto `ax` pode vir como uma matriz 2D (por exemplo, shape (2, 2)).
# Para facilitar a itera√ß√£o linear, transformamos `ax` em um array 1D com `ravel()`.
# Assim, podemos acessar o i-√©simo eixo diretamente por `ax[i]`.
ax = ax.ravel()

# Loop principal: percorre a lista de m√©tricas e cria um gr√°fico de barras para cada uma.
# - `enumerate(metricas)` devolve pares (i, metrica),
#   onde `i` √© o √≠ndice (0, 1, 2, ...) e `metrica` √© o nome da coluna.
for i, metrica in enumerate(metricas):
    # Antes de plotar, ordenamos o DataFrame `tabela_media_classe_trabalho`
    # com base na m√©trica atual.
    # - `sort_values(metrica, ascending=False)` ordena da maior para a menor m√©dia.
    # Isso deixa o gr√°fico mais intuitivo: barras maiores aparecem no topo.
    dados_plot = tabela_media_classe_trabalho.sort_values(metrica, ascending=False)

    # Cria o gr√°fico de barras horizontal (barplot) usando Seaborn.
    # - x=metrica: coloca os valores da m√©trica no eixo x (comprimento das barras).
    # - y='classe_trabalho': usa as categorias de classe de trabalho no eixo y.
    # - data=dados_plot: DataFrame com os dados j√° ordenados.
    # - ax=ax[i]: desenha o gr√°fico no eixo correspondente ao √≠ndice i.
    sns.barplot(
        x=metrica,
        y='classe_trabalho',
        data=dados_plot,
        ax=ax[i]
    )

    # Define o t√≠tulo do subplot atual usando a lista `titulos`.
    # - `titulos[i]` √© um texto descritivo que corresponde √† m√©trica `metricas[i]`.
    ax[i].set_title(titulos[i])

    # Define o r√≥tulo do eixo x.
    # - `metrica.replace('_', ' ')` substitui os underscores por espa√ßos,
    #   deixando o texto mais leg√≠vel.
    # - `.capitalize()` coloca a primeira letra em mai√∫scula.
    ax[i].set_xlabel(metrica.replace('_', ' ').capitalize())

    # Define o r√≥tulo do eixo y como "Classe de trabalho",
    # pois todas as barras representam categorias dessa vari√°vel.
    ax[i].set_ylabel('Classe de trabalho')

    # Segundo loop: adiciona os valores num√©ricos ao lado de cada barra no gr√°fico.
    # - `enumerate(dados_plot[metrica])` percorre os valores da m√©trica,
    #   j√° na ordem em que foram plotados (do maior para o menor).
    # - `idx` √© a posi√ß√£o da barra no eixo y,
    #   `value` √© o valor num√©rico da m√©trica para aquela classe de trabalho.
    for idx, value in enumerate(dados_plot[metrica]):
        ax[i].text(
            value,          # posi√ß√£o x do texto (na ponta da barra)
            idx,            # posi√ß√£o y do texto (√≠ndice da barra)
            f'{value:.2f}', # texto exibido: valor formatado com 2 casas decimais
            color='black',  # cor do texto
            va='center',    # alinhamento vertical (centralizado em rela√ß√£o √† barra)
            ha='left'       # alinhamento horizontal (texto √† direita da ponta da barra)
        )

# Ap√≥s criar todos os gr√°ficos, pode sobrar algum eixo "vazio"
# se `row * col` (total de posi√ß√µes no grid) for maior que `len(metricas)`.
# Esse loop remove os eixos que n√£o foram usados.
# - `range(i + 1, row * col)` come√ßa no pr√≥ximo √≠ndice ap√≥s o √∫ltimo gr√°fico (i)
#   e vai at√© o total de subplots dispon√≠veis, apagando os excedentes.
for j in range(i + 1, row * col):
    fig.delaxes(ax[j])

# Ajusta automaticamente o espa√ßamento entre os subplots,
# evitando sobreposi√ß√£o de t√≠tulos, r√≥tulos de eixos e valores de texto.
plt.tight_layout()

# Exibe a figura com todos os subplots.
plt.show()
```

```python
# Tabela com m√©dias por escolaridade

# Cria uma tabela resumo com as m√©dias das vari√°veis num√©ricas
# para cada categoria de 'escolaridade'
tabela_media_escolaridade = (
    dataset
        # 1) Agrupa o DataFrame pela vari√°vel categ√≥rica 'escolaridade'
        # ------------------------------------------------------------------
        # `.groupby('escolaridade', as_index=False)` informa ao pandas que:
        #   - queremos separar o DataFrame original `dataset` em grupos,
        #     onde cada grupo cont√©m todas as linhas com o mesmo valor
        #     na coluna 'escolaridade'.
        #   - `as_index=False` garante que a coluna 'escolaridade' continue
        #     sendo uma coluna normal no resultado, em vez de virar √≠ndice.
        .groupby('escolaridade', as_index=False)

        # 2) Calcula a M√âDIA das vari√°veis num√©ricas selecionadas
        # ------------------------------------------------------------------
        # Para cada n√≠vel de escolaridade (cada grupo formado acima),
        # ser√£o calculadas as seguintes m√©dias:
        #   - 'horas_semanais'       ‚Üí m√©dia de horas trabalhadas por semana
        #   - 'escolaridade_anos'    ‚Üí m√©dia de anos de estudo
        #   - 'ganho_capital'        ‚Üí m√©dia de ganhos de capital
        #   - 'perda_capital'        ‚Üí m√©dia de perdas de capital
        #
        # O resultado desta etapa √© um novo DataFrame em que:
        #   - cada linha representa uma categoria de escolaridade;
        #   - as colunas num√©ricas representam as m√©dias calculadas
        #     para aquela categoria.
        .agg({
            'horas_semanais': 'mean',
            'escolaridade_anos': 'mean',
            'ganho_capital': 'mean',
            'perda_capital': 'mean'
        })

        # 3) Arredonda todas as colunas num√©ricas para 2 casas decimais
        # ------------------------------------------------------------------
        # `.round(2)` aplica arredondamento com 2 casas decimais em todas
        # as colunas num√©ricas do DataFrame resultante.
        # Isso deixa a tabela mais limpa e amig√°vel para visualiza√ß√£o
        # (por exemplo, em relat√≥rios, dashboards ou an√°lise explorat√≥ria).
        .round(2)

        # 4) Renomeia as colunas para deixar claro que s√£o M√âDIAS
        # ------------------------------------------------------------------
        # `.rename(columns={...})` altera os nomes das colunas, deixando
        # expl√≠cito no pr√≥prio nome que os valores s√£o m√©dias agregadas,
        # e n√£o valores individuais.
        .rename(columns={
            'horas_semanais': 'media_horas_semanais',        # m√©dia de horas semanais
            'escolaridade_anos': 'media_escolaridade_anos',  # m√©dia de anos de estudo
            'ganho_capital': 'media_ganho_capital',          # m√©dia de ganho de capital
            'perda_capital': 'media_perda_capital'           # m√©dia de perda de capital
        })

        # 5) Ordena pela m√©dia de anos de estudo (tempo de estudo)
        # ------------------------------------------------------------------
        # `.sort_values('media_escolaridade_anos', ascending=False)`
        # ordena as linhas da tabela da maior para a menor m√©dia de anos
        # de escolaridade.
        # Dessa forma, as categorias com maior n√≠vel m√©dio de estudo
        # aparecem primeiro, facilitando a compara√ß√£o entre grupos.
        .sort_values('media_escolaridade_anos', ascending=False)

        # `reset_index(drop=True)` reorganiza o √≠ndice ap√≥s a ordena√ß√£o.
        # ------------------------------------------------------------------
        # - Cria um novo √≠ndice sequencial (0, 1, 2, ...) para o DataFrame.
        # - `drop=True` indica que o √≠ndice antigo n√£o deve ser mantido
        #   como uma coluna adicional.
        .reset_index(drop=True)
)

# Exibe a tabela resumo com as m√©dias por escolaridade
# (j√° arredondadas, renomeadas e ordenadas).
tabela_media_escolaridade

```

```python
import math
import matplotlib.pyplot as plt
import seaborn as sns

# Configura o estilo visual padr√£o dos gr√°ficos.
# Aqui usamos o tema "whitegrid", que deixa o fundo branco com linhas de grade,
# facilitando a leitura e a compara√ß√£o dos valores nos gr√°ficos de barras.
sns.set(style="whitegrid")

# Lista com os nomes das colunas num√©ricas (m√©tricas) que queremos visualizar.
# Cada uma dessas colunas deve estar presente no DataFrame `tabela_media_escolaridade`,
# que j√° cont√©m as m√©dias calculadas por n√≠vel de escolaridade.
metricas = [
    'media_horas_semanais',      # m√©dia de horas trabalhadas por semana
    'media_escolaridade_anos',   # m√©dia de anos de escolaridade
    'media_ganho_capital',       # m√©dia de ganho de capital
    'media_perda_capital'        # m√©dia de perda de capital
]

# Lista de t√≠tulos para cada subplot (cada gr√°fico de barras).
# A posi√ß√£o de cada t√≠tulo nesta lista corresponde √† posi√ß√£o da m√©trica em `metricas`.
# Ou seja, titulos[0] descreve o gr√°fico que usa metricas[0], e assim por diante.
titulos = [
    'M√©dia de horas semanais por escolaridade',
    'M√©dia de anos de escolaridade por escolaridade',
    'M√©dia de ganho de capital por escolaridade',
    'M√©dia de perda de capital por escolaridade'
]

# Define a quantidade de colunas do grid de subplots.
# Teremos 2 gr√°ficos por linha.
col = 2

# Calcula dinamicamente o n√∫mero de linhas necess√°rias para acomodar
# todos os gr√°ficos, dado o n√∫mero de m√©tricas e o n√∫mero de colunas.
# - len(metricas): quantidade total de gr√°ficos que ser√£o criados.
# - Dividimos por `col` para saber quantas linhas seriam necess√°rias.
# - math.ceil(...): arredonda para cima, garantindo linhas suficientes
#   mesmo que a divis√£o n√£o seja exata (por exemplo, 3 gr√°ficos em 2 colunas ‚Üí 2 linhas).
row = math.ceil(len(metricas) / col)

# Cria a figura (`fig`) e os eixos (`ax`) onde os subplots ser√£o desenhados.
# - `row`: n√∫mero de linhas no grid.
# - `col`: n√∫mero de colunas no grid.
# - `figsize=(22, 8)`: tamanho da figura em polegadas (largura x altura),
#   para garantir boa legibilidade dos textos e barras.
fig, ax = plt.subplots(row, col, figsize=(22, 8))

# O objeto `ax` pode ser um array 2D (por exemplo, 2x2).
# Para facilitar o loop, transformamos `ax` em um array 1D com `ravel()`,
# assim podemos acessar diretamente o i-√©simo eixo como `ax[i]`.
ax = ax.ravel()

# Loop principal: percorre a lista de m√©tricas e cria um gr√°fico de barras para cada uma.
# - `enumerate(metricas)` retorna pares (i, metrica),
#   onde `i` √© o √≠ndice do gr√°fico (0, 1, 2, ...) e `metrica` √© o nome da coluna num√©rica.
for i, metrica in enumerate(metricas):
    # Antes de plotar, ordenamos o DataFrame `tabela_media_escolaridade`
    # com base na m√©trica atual.
    # - `sort_values(metrica, ascending=False)` organiza as categorias de escolaridade
    #   da maior m√©dia para a menor, deixando o gr√°fico mais intuitivo visualmente.
    dados_plot = tabela_media_escolaridade.sort_values(metrica, ascending=False)

    # Cria um gr√°fico de barras horizontal (barplot) usando Seaborn.
    # - x=metrica: comprimento das barras ser√° o valor da m√©trica.
    # - y='escolaridade': cada barra representa uma categoria de escolaridade.
    # - data=dados_plot: DataFrame com os dados ordenados.
    # - ax=ax[i]: desenha o gr√°fico no eixo correspondente ao √≠ndice i.
    sns.barplot(
        x=metrica,
        y='escolaridade',
        data=dados_plot,
        ax=ax[i]
    )

    # Define o t√≠tulo do gr√°fico atual usando a lista `titulos`.
    # - titulos[i] √© um texto descritivo que explica o que est√° sendo mostrado.
    ax[i].set_title(titulos[i])

    # Define o r√≥tulo (nome) do eixo x.
    # - `metrica.replace('_', ' ')` substitui underscores por espa√ßos
    #   para ficar mais leg√≠vel.
    # - `.capitalize()` deixa a primeira letra em mai√∫scula.
    ax[i].set_xlabel(metrica.replace('_', ' ').capitalize())

    # Define o r√≥tulo do eixo y como "Escolaridade",
    # j√° que o eixo y representa os diferentes n√≠veis de escolaridade.
    ax[i].set_ylabel('Escolaridade')

    # Segundo loop: adiciona o valor num√©rico ao lado de cada barra,
    # para deixar o gr√°fico mais informativo.
    # - `enumerate(dados_plot[metrica])` percorre os valores da m√©trica
    #   na mesma ordem em que foram plotados (j√° ordenados).
    # - `idx` √© a posi√ß√£o da barra no eixo y (0, 1, 2, ...).
    # - `value` √© o valor num√©rico da m√©trica para aquela categoria.
    for idx, value in enumerate(dados_plot[metrica]):
        ax[i].text(
            value,          # posi√ß√£o x do texto (na ponta da barra)
            idx,            # posi√ß√£o y do texto (linha correspondente √† barra)
            f'{value:.2f}', # texto que ser√° exibido, formatado com 2 casas decimais
            color='black',  # cor do texto
            va='center',    # alinhamento vertical (centralizado na altura da barra)
            ha='left'       # alinhamento horizontal (texto √† direita da ponta da barra)
        )

# Depois de criar todos os gr√°ficos, pode acontecer de sobrar algum eixo
# se `row * col` (total de posi√ß√µes no grid) for maior que `len(metricas)`.
# Este loop remove os eixos que n√£o foram utilizados (subplots vazios).
# - `range(i + 1, row * col)` come√ßa no pr√≥ximo √≠ndice ap√≥s o √∫ltimo gr√°fico criado
#   e vai at√© o n√∫mero total de eixos dispon√≠veis.
for j in range(i + 1, row * col):
    fig.delaxes(ax[j])

# Ajusta automaticamente o espa√ßamento entre os subplots,
# evitando sobreposi√ß√£o de t√≠tulos, r√≥tulos e n√∫meros.
plt.tight_layout()

# Exibe a figura final com todos os gr√°ficos na tela.
plt.show()

```

```python
import math
import matplotlib.pyplot as plt
import seaborn as sns

# Configura o estilo visual padr√£o dos gr√°ficos do Seaborn/Matplotlib.
# Aqui escolhemos "whitegrid", que adiciona um fundo branco com grades,
# facilitando a leitura dos valores e compara√ß√£o entre categorias.
sns.set(style="whitegrid")

# Lista com os nomes das colunas (m√©tricas num√©ricas) que ser√£o usadas nos gr√°ficos.
# Cada string aqui corresponde a uma coluna do DataFrame `tabela_media_ocupacao`,
# que cont√©m as m√©dias calculadas previamente por ocupa√ß√£o.
metricas = [
    'media_horas_semanais',      # m√©dia de horas trabalhadas por semana
    'media_escolaridade_anos',   # m√©dia de anos de estudo
    'media_ganho_capital',       # m√©dia de ganho de capital
    'media_perda_capital'        # m√©dia de perda de capital
]

# Lista de t√≠tulos que ser√£o usados em cada subplot.
# A posi√ß√£o de cada t√≠tulo nesta lista corresponde √† posi√ß√£o da m√©trica em `metricas`.
# Assim, titulos[i] √© o t√≠tulo do gr√°fico que plota metricas[i].
# Obs.: aqui o texto ainda menciona "Estado Civil", mas os dados usados s√£o por ocupa√ß√£o.
titulos = [
    'M√©dia de horas semanais por Estado Civil',
    'M√©dia de anos de escolaridade por Estado Civil',
    'M√©dia de ganho de capital por Estado Civil',
    'M√©dia de perda de capital por Estado Civil'
]

# Define a quantidade de colunas do grid de subplots.
# Ou seja, teremos 2 gr√°ficos por linha.
col = 2

# Calcula dinamicamente a quantidade de linhas necess√°rias no grid de subplots.
# - len(metricas) retorna o n√∫mero total de gr√°ficos que queremos criar.
# - Dividimos pelo n√∫mero de colunas (`col`) para ver quantas linhas seriam necess√°rias.
# - math.ceil(...) arredonda para cima, garantindo que haja linhas suficientes
#   mesmo se a divis√£o n√£o for exata (por exemplo, 3 gr√°ficos com 2 colunas ‚Üí 2 linhas).
row = math.ceil(len(metricas) / col)

# Cria a figura (`fig`) e a matriz de eixos (`ax`) para os subplots.
# - `row` √© o n√∫mero de linhas de subplots.
# - `col` √© o n√∫mero de colunas de subplots.
# - `figsize=(22, 12)` define o tamanho da figura em polegadas (largura x altura)
#   para que todos os gr√°ficos fiquem leg√≠veis, mesmo com muitos r√≥tulos no eixo y.
fig, ax = plt.subplots(row, col, figsize=(22, 12))

# `ax` pode ser uma matriz 2D (por exemplo, shape (2,2)) dependendo de `row` e `col`.
# Para facilitar a itera√ß√£o, transformamos `ax` em um array 1D com `ravel()`.
# Assim, podemos acessar o i-√©simo eixo diretamente como `ax[i]`.
ax = ax.ravel()

# Loop principal: percorre a lista de m√©tricas e cria um gr√°fico de barras para cada uma.
# - `enumerate(metricas)` retorna pares (i, metrica),
#   onde `i` √© o √≠ndice (0, 1, 2, ...) e `metrica` √© o nome da coluna num√©rica.
for i, metrica in enumerate(metricas):
    # Antes de criar o gr√°fico, ordenamos o DataFrame `tabela_media_ocupacao`
    # com base na m√©trica atual.
    # - `sort_values(metrica, ascending=False)` ordena do maior valor para o menor,
    #   o que melhora a visualiza√ß√£o (barra maior no topo).
    dados_plot = tabela_media_ocupacao.sort_values(metrica, ascending=False)

    # Cria um gr√°fico de barras horizontal (barplot) usando Seaborn.
    # - x=metrica: valores num√©ricos no eixo x (comprimento das barras).
    # - y='ocupacao': categorias no eixo y (uma barra para cada ocupa√ß√£o).
    # - data=dados_plot: DataFrame com as colunas usadas.
    # - ax=ax[i]: desenha o gr√°fico no eixo correspondente ao √≠ndice i.
    sns.barplot(
        x=metrica,
        y='ocupacao',
        data=dados_plot,
        ax=ax[i]
    )

    # Define o t√≠tulo do subplot atual usando a lista `titulos`.
    # titulos[i] corresponde semanticamente √† m√©trica `metricas[i]`.
    # (apesar do texto mencionar "Estado Civil", aqui o recorte √© por ocupa√ß√£o).
    ax[i].set_title(titulos[i])

    # Define o r√≥tulo do eixo x.
    # - `metrica.replace('_', ' ')` troca underscores por espa√ßos.
    # - `.capitalize()` coloca a primeira letra em mai√∫scula.
    # Isso gera um r√≥tulo mais leg√≠vel a partir do nome da coluna.
    ax[i].set_xlabel(metrica.replace('_', ' ').capitalize())

    # Define o r√≥tulo do eixo y como "Ocupa√ß√£o",
    # pois todas as barras representam categorias de ocupa√ß√£o.
    ax[i].set_ylabel('Ocupa√ß√£o')

    # Segundo loop: adiciona os valores num√©ricos ao lado de cada barra do gr√°fico.
    # - `enumerate(dados_plot[metrica])` percorre os valores da m√©trica
    #   j√° na ordem em que foram plotados (do maior para o menor).
    # - `idx` √© a posi√ß√£o da barra no eixo y, `value` √© o valor num√©rico da m√©trica.
    for idx, value in enumerate(dados_plot[metrica]):
        ax[i].text(
            value,          # posi√ß√£o x onde o texto ser√° inserido (na ponta da barra)
            idx,            # posi√ß√£o y correspondente √† barra (√≠ndice da linha)
            f'{value:.2f}', # texto mostrado: valor formatado com 2 casas decimais
            color='black',  # cor do texto
            va='center',    # alinhamento vertical do texto (center: centralizado na barra)
            ha='left'       # alinhamento horizontal (left: texto √† direita da ponta da barra)
        )

# Ap√≥s criar todos os gr√°ficos, pode sobrar algum "eixo vazio"
# se o n√∫mero de m√©tricas n√£o preencher totalmente o grid (row * col).
# Esse loop remove eixos n√£o usados.
# - `range(i + 1, row * col)` come√ßa do pr√≥ximo √≠ndice ap√≥s o √∫ltimo gr√°fico usado (i)
#   e vai at√© o total de eixos (row * col) para apagar os excedentes.
for j in range(i + 1, row * col):
    fig.delaxes(ax[j])

# Ajusta automaticamente o espa√ßamento entre subplots para evitar sobreposi√ß√£o
# de t√≠tulos, r√≥tulos e n√∫meros. Deixa tudo mais "organizado" na figura final.
plt.tight_layout()

# Exibe a figura com todos os subplots na tela.
plt.show()
```

```python
# A vari√°vel `tabela_media_pais_origem` vai armazenar um novo DataFrame
# constru√≠do a partir do DataFrame original `dataset`.
# Este novo DataFrame trar√°, para cada pa√≠s de origem (`pais_origem`),
# as m√©dias das principais vari√°veis num√©ricas de interesse.
tabela_media_pais_origem = (
    dataset
        # 1) Agrupa o DataFrame pela vari√°vel categ√≥rica 'pais_origem'
        # ------------------------------------------------------------------
        # - `.groupby('pais_origem', as_index=False)` diz ao pandas:
        #     "Quero juntar as linhas que t√™m o mesmo valor em 'pais_origem'."
        # - Cada grupo vai representar um pa√≠s de origem (por exemplo:
        #     'United-States', 'Mexico', 'Philippines', etc., conforme o dataset).
        # - `as_index=False` faz com que a coluna 'pais_origem' continue sendo
        #   uma coluna normal no resultado (e n√£o vire √≠ndice do DataFrame).
        .groupby('pais_origem', as_index=False)

        # 2) Calcula a M√âDIA das vari√°veis num√©ricas selecionadas
        # ------------------------------------------------------------------
        # - `.agg({...})` aplica fun√ß√µes de agrega√ß√£o para cada grupo
        #   definido no groupby (ou seja, para cada pa√≠s de origem).
        # - Para cada categoria de 'pais_origem', ser√° calculado:
        #     * a m√©dia de 'horas_semanais'       ‚Üí jornada m√©dia de trabalho
        #     * a m√©dia de 'escolaridade_anos'    ‚Üí m√©dia de anos de estudo
        #     * a m√©dia de 'ganho_capital'        ‚Üí ganho m√©dio de capital
        #     * a m√©dia de 'perda_capital'        ‚Üí perda m√©dia de capital
        # - O resultado √© um DataFrame onde cada linha representa
        #   um pa√≠s de origem e as colunas num√©ricas representam as m√©dias.
        .agg({
            'horas_semanais': 'mean',        # m√©dia de horas trabalhadas por semana
            'escolaridade_anos': 'mean',     # m√©dia de anos de estudo
            'ganho_capital': 'mean',         # m√©dia de ganhos de capital
            'perda_capital': 'mean'          # m√©dia de perdas de capital
        })

        # 3) Arredonda todas as colunas num√©ricas para 2 casas decimais
        # ------------------------------------------------------------------
        # - `.round(2)` aplica o arredondamento em todas as colunas
        #   de tipo num√©rico do DataFrame resultante.
        # - Isso deixa a tabela mais "limpa" para apresenta√ß√£o (relat√≥rios,
        #   gr√°ficos, etc.), evitando muitos d√≠gitos decimais.
        .round(2)

        # 4) Renomeia as colunas para deixar claro que s√£o M√âDIAS
        # ------------------------------------------------------------------
        # - `.rename(columns={...})` altera os nomes das colunas.
        # - Aqui, deixamos expl√≠cito no nome de cada vari√°vel que o valor
        #   representa uma m√©dia, n√£o um valor individual.
        .rename(columns={
            'horas_semanais': 'media_horas_semanais',
            'escolaridade_anos': 'media_escolaridade_anos',
            'ganho_capital': 'media_ganho_capital',
            'perda_capital': 'media_perda_capital'
        })

        # 5) Ordena pela m√©dia de anos de estudo (tempo de estudo)
        # ------------------------------------------------------------------
        # - `.sort_values('media_escolaridade_anos', ascending=False)`
        #   ordena os pa√≠ses de origem da maior m√©dia de escolaridade para a menor.
        # - Isso facilita identificar rapidamente quais pa√≠ses de origem
        #   concentram indiv√≠duos com mais anos m√©dios de estudo.
        .sort_values('media_escolaridade_anos', ascending=False)

        # 6) Reorganiza o √≠ndice do DataFrame
        # ------------------------------------------------------------------
        # - `.reset_index(drop=True)` cria um novo √≠ndice sequencial
        #   (0, 1, 2, ...) ap√≥s a ordena√ß√£o.
        # - `drop=True` indica que o √≠ndice antigo n√£o deve ser mantido
        #   como uma coluna adicional.
        .reset_index(drop=True)
)

# Exibe a tabela resultante com as m√©dias por pa√≠s de origem,
# j√° arredondadas, renomeadas e ordenadas.
tabela_media_pais_origem
```

```python
import math
import matplotlib.pyplot as plt
import seaborn as sns

# Configura o estilo visual padr√£o dos gr√°ficos do Seaborn/Matplotlib.
# Aqui escolhemos "whitegrid", que adiciona um fundo branco com grades,
# facilitando a leitura dos valores e a compara√ß√£o entre categorias.
sns.set(style="whitegrid")

# Lista com os nomes das colunas (m√©tricas num√©ricas) que ser√£o usadas nos gr√°ficos.
# Cada string aqui corresponde a uma coluna do DataFrame `tabela_media_pais_origem`,
# que cont√©m as m√©dias calculadas previamente por pa√≠s de origem.
metricas = [
    'media_horas_semanais',      # m√©dia de horas trabalhadas por semana
    'media_escolaridade_anos',   # m√©dia de anos de estudo
    'media_ganho_capital',       # m√©dia de ganho de capital
    'media_perda_capital'        # m√©dia de perda de capital
]

# Lista de t√≠tulos que ser√£o usados em cada subplot.
# A posi√ß√£o de cada t√≠tulo nesta lista corresponde √† posi√ß√£o da m√©trica em `metricas`.
# Assim, titulos[i] √© o t√≠tulo do gr√°fico que plota metricas[i].
titulos = [
    'M√©dia de horas semanais por pa√≠s de origem',
    'M√©dia de anos de escolaridade por pa√≠s de origem',
    'M√©dia de ganho de capital por pa√≠s de origem',
    'M√©dia de perda de capital por pa√≠s de origem'
]

# Define a quantidade de colunas do grid de subplots.
# Ou seja, teremos 2 gr√°ficos por linha.
col = 2

# Calcula dinamicamente a quantidade de linhas necess√°rias no grid de subplots.
# - len(metricas) retorna o n√∫mero total de gr√°ficos que queremos criar.
# - Dividimos pelo n√∫mero de colunas (`col`) para ver quantas linhas seriam necess√°rias.
# - math.ceil(...) arredonda para cima, garantindo que haja linhas suficientes
#   mesmo se a divis√£o n√£o for exata (por exemplo, 3 gr√°ficos com 2 colunas ‚Üí 2 linhas).
row = math.ceil(len(metricas) / col)

# Cria a figura (`fig`) e a matriz de eixos (`ax`) para os subplots.
# - `row` √© o n√∫mero de linhas de subplots.
# - `col` √© o n√∫mero de colunas de subplots.
# - `figsize=(22, 12)` define o tamanho da figura em polegadas (largura x altura)
#   para que todos os gr√°ficos fiquem leg√≠veis, mesmo com muitos r√≥tulos no eixo y.
fig, ax = plt.subplots(row, col, figsize=(25, 22))

# `ax` pode ser uma matriz 2D (por exemplo, shape (2,2)) dependendo de `row` e `col`.
# Para facilitar a itera√ß√£o, transformamos `ax` em um array 1D com `ravel()`.
# Assim, podemos acessar o i-√©simo eixo diretamente como `ax[i]`.
ax = ax.ravel()

# Loop principal: percorre a lista de m√©tricas e cria um gr√°fico de barras para cada uma.
# - `enumerate(metricas)` retorna pares (i, metrica),
#   onde `i` √© o √≠ndice (0, 1, 2, ...) e `metrica` √© o nome da coluna num√©rica.
for i, metrica in enumerate(metricas):
    # Antes de criar o gr√°fico, ordenamos o DataFrame `tabela_media_pais_origem`
    # com base na m√©trica atual.
    # - `sort_values(metrica, ascending=False)` ordena do maior valor para o menor,
    #   o que melhora a visualiza√ß√£o (barra maior no topo).
    dados_plot = tabela_media_pais_origem.sort_values(metrica, ascending=False)

    # Cria um gr√°fico de barras horizontal (barplot) usando Seaborn.
    # - x=metrica: valores num√©ricos no eixo x (comprimento das barras).
    # - y='pais_origem': categorias no eixo y (uma barra para cada pa√≠s de origem).
    # - data=dados_plot: DataFrame com as colunas usadas.
    # - ax=ax[i]: desenha o gr√°fico no eixo correspondente ao √≠ndice i.
    sns.barplot(
        x=metrica,
        y='pais_origem',
        data=dados_plot,
        ax=ax[i]
    )

    # Define o t√≠tulo do subplot atual usando a lista `titulos`.
    # titulos[i] corresponde semanticamente √† m√©trica `metricas[i]`.
    ax[i].set_title(titulos[i])

    # Define o r√≥tulo do eixo x.
    # - `metrica.replace('_', ' ')` troca underscores por espa√ßos.
    # - `.capitalize()` coloca a primeira letra em mai√∫scula.
    # Isso gera um r√≥tulo mais leg√≠vel a partir do nome da coluna.
    ax[i].set_xlabel(metrica.replace('_', ' ').capitalize())

    # Define o r√≥tulo do eixo y como "Pa√≠s de origem",
    # pois todas as barras representam categorias de pa√≠s.
    ax[i].set_ylabel('Pa√≠s de origem')

    # Segundo loop: adiciona os valores num√©ricos ao lado de cada barra do gr√°fico.
    # - `enumerate(dados_plot[metrica])` percorre os valores da m√©trica
    #   j√° na ordem em que foram plotados (do maior para o menor).
    # - `idx` √© a posi√ß√£o da barra no eixo y, `value` √© o valor num√©rico da m√©trica.
    for idx, value in enumerate(dados_plot[metrica]):
        ax[i].text(
            value,          # posi√ß√£o x onde o texto ser√° inserido (na ponta da barra)
            idx,            # posi√ß√£o y correspondente √† barra (√≠ndice da linha)
            f'{value:.2f}', # texto mostrado: valor formatado com 2 casas decimais
            color='black',  # cor do texto
            va='center',    # alinhamento vertical do texto (center: centralizado na barra)
            ha='left'       # alinhamento horizontal (left: texto √† direita da ponta da barra)
        )

# Ap√≥s criar todos os gr√°ficos, pode sobrar algum "eixo vazio"
# se o n√∫mero de m√©tricas n√£o preencher totalmente o grid (row * col).
# Esse loop remove eixos n√£o usados.
# - `range(i + 1, row * col)` come√ßa do pr√≥ximo √≠ndice ap√≥s o √∫ltimo gr√°fico usado (i)
#   e vai at√© o total de eixos (row * col) para apagar os excedentes.
for j in range(i + 1, row * col):
    fig.delaxes(ax[j])

# Ajusta automaticamente o espa√ßamento entre subplots para evitar sobreposi√ß√£o
# de t√≠tulos, r√≥tulos e n√∫meros. Deixa tudo mais "organizado" na figura final.
plt.tight_layout()

# Exibe a figura com todos os subplots na tela.
plt.show()
```
