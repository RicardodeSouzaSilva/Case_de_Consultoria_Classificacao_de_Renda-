```python
import importlib    # Importa o m√≥dulo 'importlib', que permite importar bibliotecas dinamicamente em tempo de execu√ß√£o
import subprocess   # Importa o m√≥dulo 'subprocess' para executar comandos externos, como instalar pacotes via pip
import sys          # Importa o m√≥dulo 'sys' para acessar o interpretador Python atual e seu execut√°vel


# Dicion√°rio contendo os pacotes que devem ser verificados e instalados, se necess√°rio.
# A chave √© o nome usado para importar o m√≥dulo, e o valor √© o nome usado no pip para instala√ß√£o.
required_packages = {
    'sklearn': 'scikit-learn',     # sklearn √© importado como m√≥dulo, mas instalado com o nome scikit-learn
    'scipy': 'scipy',              # scipy √© tanto o nome de importa√ß√£o quanto de instala√ß√£o
    'pandas': 'pandas',            # pandas para an√°lise de dados tabulares
    'numpy': 'numpy',              # numpy para opera√ß√µes num√©ricas
    'matplotlib': 'matplotlib',    # matplotlib para visualiza√ß√µes
    'seaborn': 'seaborn',          # seaborn para gr√°ficos estat√≠sticos
    'missingno': 'missingno'       # missingno para visualiza√ß√£o de dados ausentes
}

# Fun√ß√£o respons√°vel por verificar se o pacote est√° instalado.
# Caso n√£o esteja, ela o instala automaticamente usando pip.
def check_and_install(package_name, install_name):
    try:
        # Tenta importar o pacote. Se for bem-sucedido, ele j√° est√° instalado.
        importlib.import_module(package_name)
        print(f"‚úÖ Biblioteca '{package_name}' j√° est√° instalada.")
    except ImportError:
        # Caso a importa√ß√£o falhe, significa que a biblioteca n√£o est√° instalada.
        # O c√≥digo abaixo usa o subprocess para executar o comando: python -m pip install <nome_do_pacote>
        print(f"‚ö†Ô∏è Biblioteca '{package_name}' n√£o encontrada. Instalando '{install_name}'...")
        subprocess.check_call([sys.executable, "-m", "pip", "install", install_name])
        print(f"‚úÖ Biblioteca '{install_name}' instalada com sucesso.")

# Loop que percorre cada item do dicion√°rio de pacotes.
# Para cada pacote, chama a fun√ß√£o de verifica√ß√£o e instala√ß√£o, se necess√°rio.
for package, install_name in required_packages.items():
    check_and_install(package, install_name)
```

### 4.2.5 Verifica√ß√£o Valores Outliers e Quartis (horas_semanais)

```python
# Calcula o primeiro quartil (Q1), terceiro quartil (Q3) e a amplitude interquartil (IQR) da coluna 'horas_semanais' do DataFrame 'dataset'
Q1_horas_semanais = dataset['horas_semanais'].quantile(0.25)  # Primeiro quartil (25%)
Q3_horas_semanais = dataset['horas_semanais'].quantile(0.75)  # Terceiro quartil (75%)
IQR_horas_semanais = Q3_horas_semanais - Q1_horas_semanais # Amplitude interquartil
```

```python
# Calcula os limites inferior e superior para identifica√ß√£o de outliers usando a regra do IQR
limite_inferior_horas_semanais = Q1_horas_semanais - 1.5 * IQR_horas_semanais  # Limite inferior
limite_superior_horas_semanais = Q3_horas_semanais + 1.5 * IQR_horas_semanais # Limite superior
```

```python
# Identificando os outliers
# Seleciona os outliers da coluna 'Renda' com base nos limites inferior e superior calculados anteriormente
outliers_horas_semanais = dataset[(dataset['horas_semanais'] < limite_inferior_horas_semanais) | (dataset['horas_semanais'] > limite_superior_horas_semanais)]  
outliers_horas_semanais
```

```python
# Tabela com m√©dias por classe de trabalho
# Cria uma tabela resumo com as m√©dias por classe de trabalho
tabela_media_classe_trabalho = (
    dataset
        # 1) Agrupa o DataFrame pela vari√°vel qualitativa 'classe_trabalho'.
        #    Cada categoria de classe_trabalho vira um grupo:
        #    ex.: 'Private', 'Self-emp-not-inc', 'Local-gov', etc.
        .groupby('classe_trabalho', as_index=False)

        # 2) Para cada grupo, calcula a M√âDIA das vari√°veis num√©ricas selecionadas.
        #    Ou seja: para cada classe_trabalho, voc√™ obt√©m:
        #      - m√©dia de horas_semanais
        #      - m√©dia de escolaridade_anos
        #      - m√©dia de ganho_capital
        #      - m√©dia de perda_capital
        .agg({
            'horas_semanais': 'mean',
            'escolaridade_anos': 'mean',
            'ganho_capital': 'mean',
            'perda_capital': 'mean'
        })

        # 3) Arredonda todas as colunas num√©ricas para 2 casas decimais,
        #    deixando a tabela mais leg√≠vel para an√°lise e apresenta√ß√£o.
        .round(2)

        # 4) Renomeia as colunas de m√©dias para nomes mais descritivos,
        #    indicando explicitamente que se trata de "m√©dia" de cada vari√°vel.
        .rename(columns={
            'horas_semanais': 'media_horas_semanais',
            'escolaridade_anos': 'media_escolaridade_anos',
            'ganho_capital': 'media_ganho_capital',
            'perda_capital': 'media_perda_capital'
        })
)

tabela_media_classe_trabalho
```

```python
import math
import matplotlib.pyplot as plt
import seaborn as sns

# Deixa o estilo mais agrad√°vel
sns.set(style="whitegrid")

# Lista das m√©tricas num√©ricas que queremos plotar
metricas = [
    'media_horas_semanais',
    'media_escolaridade_anos',
    'media_ganho_capital',
    'media_perda_capital'
]

# T√≠tulos bonitinhos para cada subplot
titulos = [
    'M√©dia de horas semanais por classe de trabalho',
    'M√©dia de anos de escolaridade por classe de trabalho',
    'M√©dia de ganho de capital por classe de trabalho',
    'M√©dia de perda de capital por classe de trabalho'
]

# Quantidade de colunas no grid
col = 2

# Calcula o n√∫mero de linhas dinamicamente
row = math.ceil(len(metricas) / col)

# Cria figura e eixos
fig, ax = plt.subplots(row, col, figsize=(22, 8))

# Garante que ax seja um array 1D para iterar mais f√°cil
ax = ax.ravel()

# Itera sobre as m√©tricas e cria um gr√°fico de barras para cada uma
for i, metrica in enumerate(metricas):
    # Ordena os dados pela m√©trica atual (opcional, mas deixa visual melhor)
    dados_plot = tabela_media_classe_trabalho.sort_values(metrica, ascending=False)

    # Cria o barplot horizontal
    sns.barplot(
        x=metrica,
        y='classe_trabalho',
        data=dados_plot,
        ax=ax[i]
    )

    # Define t√≠tulo e r√≥tulos de eixos
    ax[i].set_title(titulos[i])
    ax[i].set_xlabel(metrica.replace('_', ' ').capitalize())
    ax[i].set_ylabel('Classe de trabalho')

    # Adiciona o valor num√©rico ao lado de cada barra
    for idx, value in enumerate(dados_plot[metrica]):
        ax[i].text(
            value,                    # posi√ß√£o x
            idx,                      # posi√ß√£o y (√≠ndice da barra)
            f'{value:.2f}',           # texto (valor formatado)
            color='black',
            va='center',
            ha='left'
        )

# Se sobrar eixo (quando o n√∫mero de m√©tricas n√£o fecha o grid), apaga
for j in range(i + 1, row * col):
    fig.delaxes(ax[j])

# Ajusta o layout para n√£o sobrepor elementos
plt.tight_layout()

# Mostra o gr√°fico
plt.show()
```
